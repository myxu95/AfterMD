## TCR-pMHC结合示意

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(facecolor='w')
#ax = Axes3D(fig)

x = np.linspace(0.5, 3.5, 5)
y = np.linspace(0.5, 3.5, 5)
X, Y = np.meshgrid(x, y)
## np.meshgrid：生成网格点坐标矩阵（x与y中的点坐标可以随意组合）
#ax.plot_surface(X, Y, Z=X*0+1, color='g', alpha=0.6)

x0 = np.linspace(1, 3, 50)
z0 = np.linspace(0.5, 5, 50)
x1 = np.linspace(2, 3, 50)

#pMHC
ax1 = fig.add_subplot(projection='3d') 
ax1.plot_surface(X, Y, Z=X*0+1, color='w', alpha=0.6) ## 平面
l1, = ax1.plot(x0, x0*0+2, x0*0+1, color='purple', linewidth=2.5)  ## Pep
l2, =ax1.plot(z0*0+2, z0*0+2, z0, color='gold', linewidth=2) ##法线

l3, = ax1.plot(x1, 5-x1, 4, color='r', linewidth=2) ## twist
l4, = ax1.plot(x1, x1, x1*4-6, color='g', linewidth=2) ## tilt
l5, = ax1.plot(x1, x1, 5.25-x1*0.5, color='b', linewidth=2) ## swing

plt.legend(handles=[l1, l2, l3, l4, l5], labels=[ 'Pep', 'MHC normal', 'TCR twist axis' , 'TCR tilt axis', 'TCR swing axis' ], loc='best') 
## 注释

ax1.scatter(2, 2, 2,  c="black", marker="o", s=25)

axs = fig.gca()
axs.set(xlabel='X', ylabel='Y', zlabel='Z', xlim=(-2, 5), ylim=(-2, 5), zlim=(0, 6),
       xticks=np.arange(-2, 6, 1), yticks=np.arange(-2, 6, 1), zticks=np.arange(0, 7, 1))


plt.show()
```
